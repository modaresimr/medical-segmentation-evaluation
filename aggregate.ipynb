```python

results_dir = 'results'
task='Task001_LiverTumor'

import os, json,copy

m_def={'D':{'tp':0,'fp':0,'fn':0,'tn':0},
                   'B':{'tp':0,'fp':0,'fn':0,'tn':0},
                   'U':{'tp':0,'fp':0,'fn':0,'tn':0},
                   'R':{'tp':0,'fp':0,'fn':0,'tn':0},
                   'T':{'tp':0,'fp':0,'fn':0,'tn':0}
                  }
all={}
for alg in os.listdir(f'{results_dir}/{task}/'):
    
    all[alg]={}
    for item in os.listdir(f'{results_dir}/{task}/{alg}/'):
        if '.json' in item:
            with open(f'{results_dir}/{task}/{alg}/{item}','r') as f:
                data=json.load(f)
            for c in data['MME']:
#                 print(item,data['MME'][c]['total'])
                if c not in all[alg]:
                    all[alg][c]=copy.deepcopy(m_def)
                for p in all[alg][c]:
                    for d in all[alg][c][p]:
                        all[alg][c][p][d]+=data['MME'][c]['total'][p][d] if data['MME'][c]['total'][p][d]>0 else 0
            
# {alg: for alg in all}
print(all)
```
