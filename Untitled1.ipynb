```python
from skimage.morphology import skeletonize
from scipy.ndimage import distance_transform_edt, generate_binary_structure, binary_erosion
import cc3d
reference=gt
gt_labels, gN = cc3d.connected_components(reference, return_N=True) 
helper={}
helper['gt_labels']=gt_labels
helper['gN']=gN

footprint = generate_binary_structure(reference.ndim, connectivity=1)
helper['footprint']=footprint

helper['components']={}
for i in range(1,gN+1):
    gt_component= gt_labels==i
    gt_border=~binary_erosion(gt_component, structure=footprint, iterations=1)& (gt_component)
    gt_border_dst=distance_transform_edt(~gt_border)
    out_dst=distance_transform_edt(~gt_component)
    in_dst=distance_transform_edt(gt_component)

    skeleton = skeletonize(gt_component)
    skeleton_dst=distance_transform_edt(skeleton==0)


    normalize_dst_inside=in_dst/(skeleton_dst+in_dst)
    normalize_dst_outside=out_dst/(skeleton_dst-out_dst+0.0000000001)
    normalize_dst=normalize_dst_inside+normalize_dst_outside


    helper['components'][i]={
        'gt':gt_component,
        'gt_border':gt_border,
        'gt_border_dst':gt_border_dst,
        'gt_out_dst':out_dst,
        'gt_in_dst':in_dst,
        'gt_skeleton':skeleton,
        'gt_skeleton_dst':skeleton_dst,
        'normalize_dst_inside':normalize_dst_inside,
        'normalize_dst_outside':normalize_dst_outside,
        'normalize_dst':normalize_dst
    }

```
